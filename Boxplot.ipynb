# CREACIÓN DE UN BOXPLOT

## Paso 1: Importar las librerías necesarias
Primero, necesitas importar las bibliotecas que utilizaremos, que son numpy para la creación de datos y matplotlib para la visualización.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Paso 2: Generar un conjunto de datos
Creamos un conjunto de datos aleatorios con distribución normal. Usaremos numpy para hacerlo.

```python
# Crear un conjunto de datos aleatorios
np.random.seed(42)  # Para reproducibilidad
data = np.random.normal(loc=50, scale=10, size=100)  # media=50, desviación estándar=10 número valores 100
```

## Paso 3: Mostrar los datos
Vamos a ver algunos valores de nuestro conjunto de datos para asegurarnos de que se generaron correctamente.

```python
print(data[:10])  # Mostrar los primeros 10 valores, como vemos son decimales. No son números enteros.
```

## Paso 4: Calcular los cuartiles y otros estadísticos
Ahora, calculamos los cuartiles y otros estadísticos necesarios para la construcción del boxplot.

```python
Q1 = np.percentile(data, 25)  # Primer cuartil, la base de datos se llama "data" y se pide el percentil 25, y así sucesivamente.
Q2 = np.percentile(data, 50)   # Mediana
Q3 = np.percentile(data, 75)   # Tercer cuartil
IQR = Q3 - Q1                  # Rango intercuartílico
lower_bound = Q1 - 1.5 * IQR   # Límite inferior
upper_bound = Q3 + 1.5 * IQR   # Límite superior

print(f"el valor Q1 es: {Q1}, El valor Q2 es: {Q2}, El valor Q3 es: {Q3}")
print(f"El intercuartílico IQR: {IQR}, Lower bound, la línea inferior: {lower_bound}, Upper bound la línea superior: {upper_bound}")
```

## Paso 5: Crear el boxplot
Usamos matplotlib para crear y mostrar el boxplot.

```python
plt.figure(figsize=(10, 6))  # Tamaño de la figura, con un ancho de 10 y alto 6.
plt.boxplot(data, vert=True)  # Aquí se ordena que se dibuje un boxplot, vert=True para orientación vertical
plt.title("Boxplot de un conjunto de 100 valores")
plt.ylabel("Valores")
plt.grid(axis='y')
plt.show()
```

## Paso 6: Identificar y mostrar outliers
Puedes identificar los outliers basados en los límites calculados.

```python
outliers = data[(data < lower_bound) | (data > upper_bound)] # Simplemente se seleccionan los valores fuera de ese rango que hemos llamado (bound)
print(f"Sólo hay 1 Outliers: {outliers}")
```

## Lo mismo pero con un histograma

```python
plt.figure(figsize=(10, 6))                                  # La dimensión de la figura
plt.hist(data, bins=10, color='skyblue', edgecolor='black')  # bins es el número de intervalos, podemos jugar y cambiar a 20, 30... a ver qué pasa
plt.title("Histograma de un conjunto de 100 valores")        # Arriba, skyblue el color de las columnas, y el borde "black"
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(axis='y')
plt.show()
```
